```python
import ast
```

```python
def convert(var):
  list1=[]
  for i in ast.literal_eval(var) :
    list1.append(i['name'])
  return list1
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0]['cast']
```

```python
def convert2(var):
  list1=[]
  count=0
  for i in ast.literal_eval(var):
    list1.append(i['character'])
    count+=1
    if (count==5):
      break
  return list1
convert2(movies.iloc[0]['cast'])
```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
movies.head(2)
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies.iloc[1]['crew']
```

```python
def convert3(var):
  list1=[]
  for i in ast.literal_eval(var):
    if i["job"] in ['Director',"Editor","Producer"]:
        list1.append(i['name'])
  return list(set(list1))
convert3(movies.iloc[0]['crew'])
```

```python
movies['crew']= movies['crew'].apply(convert3)
```

```python
movies.head()
```

```python
 movies['overview']= movies['overview'].apply(lambda x : x.split())
```

```python
movies['genres']= movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']= movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']= movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']= movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
movies['tags']=movies['overview'] + movies['keywords'] + movies['genres'] + movies['cast'] +movies['crew']
```

```python
movies.head(1)
```

```python
df=movies[['movie_id','title','tags']]
```

```python
df.head()
```

```python
df['tags']= df['tags'].apply(lambda x :" ".join(x))
```

```python
df['tags'][1]
```

```python
df['tags']=df['tags'].apply(lambda x : x.lower())
```

```python
df['tags'][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=6000,stop_words='english')
```

```python
cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_params
```

```python
print(cv.get_feature_names_out()[:100])
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  list1=[]
  for i in text.split(" "):
    list1.append(ps.stem(i))
  return " ".join(list1)
```

```python
stem(df['tags'][1])
```

```python
df['tags']= df['tags'].apply(stem)
```

```python
vectors=cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_names_out()[:100]
```

```python
from sklearn.metrics.pairwise  import cosine_similarity
```

```python
matrix=cosine_similarity(vectors)
```

```python
matrix.shape
```

```python
#matrix[1] represents the second movie distance with all other movies
```

```python
def recommend(movie):

  movie_index=df[df['title']==movie].index[0]
  distances = matrix[movie_index]
  movies_list= sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(df.iloc[i[0]]['title'])
```

```python
recommend('Avatar')
```
